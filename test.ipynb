```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of points to generate
num_points = 1000

# Generate points for the lollipop surface (circle with radius 1 at (0, 2))
circle_x = []
circle_y = []
for _ in range(num_points):
    r = np.random.uniform(0, 1)
    r = np.sqrt(r)
    theta = np.random.uniform(0, 2 * np.pi)
    x = np.cos(theta) * r
    y = np.sin(theta) * r + 2 
    circle_x.append(x)
    circle_y.append(y)
circle_x = np.array(circle_x)
circle_y = np.array(circle_y)

# Generate points for the lollipop stick (line from (0, 0) to (0, 1))
stick_x = np.zeros(num_points)
stick_y = np.linspace(0, 1, num_points)

# Generate a set of points at (0, -1)
point_x = np.zeros(num_points)
point_y = -np.ones(num_points)

# Combine all points
x_values = np.concatenate((circle_x, stick_x, point_x))
y_values = np.concatenate((circle_y, stick_y, point_y))

# Plot the points to visualize the lollipop
plt.figure(figsize=(6, 8))
plt.scatter(x_values, y_values, s=5)
plt.axis('equal')
plt.show()
```
